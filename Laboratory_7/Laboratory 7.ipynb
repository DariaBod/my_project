#### Задание 1

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
values=ratings.groupby('userId').agg({'timestamp': ['min', 'max', 'count']})
```

```python
more_than_100=values.loc[values['timestamp']['count']>100]
more_than_100
```

```python
average_lifetime=sum(more_than_100['timestamp']['max']-more_than_100['timestamp']['min'])/len(more_than_100)
average_lifetime
```

#### Задание 3
В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
from pymystem3 import Mystem
```

```python
data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})
data
```

```python
m = Mystem()
data['lemmas'] =[' '.join(m.lemmatize(str(each))) for each in data['keyword']]
data
```
