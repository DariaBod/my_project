#### Визуализация №1
В таблице самых счастливых стран посмотреть место в рейтинге по позитивному впечатлению людей у первых 15-ти записей.

```python
from pylab import rcParams
rcParams['figure.figsize'] = 10,7

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
```

```python
countries = pd.read_csv('world-happiness-report-2019.csv')
countries_barh = countries.sort_values('Ladder').head(15)
countries_barh
```

```python
plt.barh(countries_barh['Country (region)'], countries_barh['Positive affect'])
plt.xlabel('Number in positive affect')
plt.ylabel('Countries')
plt.title('Rating of the positive affect of the happiness countries')
plt.text(4, 3, 'Most positive country')
plt.grid()
```

**Вывод:** 
Страна с самым высоким уровнем позитивного впечатления из первых пятнадцати списка самых счастливых стран это Исландия.

#### Визуализация №2
Посмотреть длину названия стран.

```python
countries['Length of country`s name'] = countries['Country (region)'].map(len)
countries['Length of country`s name'].plot(kind='box')
```

**Вывод:** самое короткое название страны состоит из 4-х символов, а самое длинное из 24-х символов.

#### Визуализация №3
По графикам посмотреть зависимость друг от друга таких парметров как Социальная поддержка, Уровень свободы и Позитивное впечатление.

```python
import seaborn as sns
```

```python
countries_15 = countries.sort_values('Ladder').head(30)
sns_plot = sns.pairplot(
    countries_15[['Social support', 'Freedom', 'Positive affect']])
sns_plot
```

**Вывод:** в большинстве стран социальная поддержка и свобода прямо пропорциональны, однако в странах, где какой-то один параметр отстаёт наблюдается обратная зависимость. Страны с меньшем уровнем социальной поддержки вызывают меньший положительный эффект.

#### Визуализация №4

#### Визуализация №5
Увидеть зависимость места в рейтинге самых счастливых стран от негативного впечатления у людей.

```python
sns.jointplot(x='Ladder', y='Negative affect', 
              data=countries, kind='reg')
```

**Вывод:** зависимость обратно пропорциональна. Чем большее место в рейтинге, тем больший уровень негативного впечатления.
